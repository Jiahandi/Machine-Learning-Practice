```python
import nn
import time
import numpy as np
import pandas as pd
%load_ext autoreload
%autoreload 2
```

```python
data_path = '../data/watermelon3.0.csv'
```

```python
data = pd.read_csv(data_path)
```

```python
y = data['label']
```

```python
y = np.array(y).reshape(1, (len(y)))
```

```python
X = data.loc[:, data.columns != 'label']
```

```python
X = np.array(X).T
```

**标准BP算法**  运行十次取收敛轮次平均值。  

```python
epoch_count = 0
time_count = 0
round_num = 25
for i in range(round_num):
    snn = nn.SimpleNeuralNetwork(X.shape[0], 5, y.shape[0])
    st = time.time()
    epoch_count += snn.train(X, y, eta=0.1, epoch_num=30000, batch_size=1)
    ed = time.time()
    time_count += ed - st
print('标准BP算法收敛需', epoch_count / round_num, 'epoch', time_count / round_num, '秒')
```

**累积BP算法** 运行十次取收敛轮次平均值。

```python
epoch_count = 0
time_count = 0
round_num = 25
for i in range(round_num):
    
    snn = nn.SimpleNeuralNetwork(X.shape[0], 5, y.shape[0])
    st = time.time()
    epoch_count += snn.train(X, y, eta=0.1, epoch_num=30000, batch_size=17)
    ed = time.time()
    time_count += ed - st
print('累积BP算法收敛需', epoch_count / round_num, 'epoch', time_count / round_num, '秒')
```
